```python
## Import packages and file

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
import seaborn as sns
%matplotlib inline


df = pd.read_csv('states_all.csv')
```

## Thinkful Unit 1 Lesson 3 Assignment 3##

```python
## A Table containing values for California Only (Stored as a Variable)
CA_isolated=df.loc[lambda df: df['STATE'] == 'CALIFORNIA', :]
```

```python
## Set Seaborn Style
sns.set(style="darkgrid")
```

**1. Choose one variable and plot that variable four different ways.**

```python
h = sns.FacetGrid(CA_isolated, height=5)
h.map(plt.plot, "YEAR", "TOTAL_REVENUE", color="steelblue")
h.set_axis_labels("Year", "Total Revenue (10's of Millions)")
h.set(ylim=(20000000, 90000000))
plt.subplots_adjust(top=0.9)
plt.title('Total Revenue of California Schools')
sns.despine(trim=True)
plt.show()
```

Pros: Easy to read, can understand trends in revenue over time
Cons: Lack of descriptive statistics

```python
h = sns.FacetGrid(CA_isolated, height=5)
h.map(plt.hist, "TOTAL_REVENUE", color="steelblue")
h.set_axis_labels("Total Revenue (Tens of Millions)", "Frequency")
plt.subplots_adjust(top=0.9)
plt.title('Total Revenue of California Schools')
sns.despine(trim=True)
plt.show()
```

Pros: Can identify where revenue Stagnated the most (however unintuitive that type of analysis would be)
Cons: Extremely difficult to make sense of

```python
sns.set(style="darkgrid")
h = sns.factorplot( y="TOTAL_REVENUE", data=CA_isolated, size=4, kind="bar", palette="pastel", ci=95)
plt.title('Total Revenue of California Schools')
h.set_axis_labels("California", "Total Revenue")
plt.show()
```

Pros: Can see the mean and variance of revenue data
Cons: Lacks context about specific datapoints, height of bar has little meaning alone

```python
sns.set(style="darkgrid")
h = sns.factorplot( y="TOTAL_REVENUE", data=CA_isolated, size=4, kind="point", palette="pastel", ci=95, dodge=True)
plt.title('Total Revenue of California Schools')
h.set_axis_labels("California", "Total Revenue")
plt.show()
```

Pros: Can see the mean and variance of revenue data, more visually appropriate than boxplot because of exclusion of unnecessary bar.
Cons: Lacks context about specific datapoints

**2. Choose two continuous variables, and plot them three different ways.**

```python
h = sns.FacetGrid(CA_isolated, height=5)
h.map(plt.scatter, "TOTAL_REVENUE", "TOTAL_EXPENDITURE", color="steelblue")
h.set_axis_labels("Total Revenue", "Total Expenditure")

plt.title('Total Revenue vs Total Expenditure')
sns.despine(trim=True)
plt.show()

```

Pros: Can see relationship between Revenue and Expenditure clearly
Cons: lacks context about individual datapoints of each variable seperately

```python
h = sns.FacetGrid(CA_isolated, height=4)
h.map(plt.stackplot, "YEAR", "TOTAL_EXPENDITURE", color="green")
h.map(plt.stackplot, "YEAR", "TOTAL_REVENUE", color="steelblue")
h.set_axis_labels("Total Revenue", "Total Expenditure")
plt.title('Total Revenue vs Total Expenditure')
sns.despine(trim=True)
plt.show()
```

Pros: It can be seen that revenue was more than expenditures for most of the years
Cons: At every year one of 2 datapoints is obscured

```python
h = sns.FacetGrid(CA_isolated, height=4)
h.map(plt.plot, "YEAR", "TOTAL_EXPENDITURE", color="steelblue")
h.map(plt.plot, "YEAR", "TOTAL_REVENUE", color="green")
h.set_axis_labels("Total Revenue", "Total Expenditure")
plt.title('Total Revenue vs Total Expenditure')
sns.despine(trim=True)
plt.show()
```

Pros: Both sets of datapoints are visible
Cons: Lack of visualization of descriptive/comparative statistics

**3. Choose one continuous variable and one categorical variable, and plot them six different ways.**

```python
## A Table containing values for Texas Only (Stored as a Variable)
TX_isolated=df.loc[lambda df: df['STATE'] == 'TEXAS', :]

## Merging CA and TX data frame
CA_TX_isolated_list =[CA_isolated, TX_isolated]
CA_TX_isolated = pd.concat(CA_TX_isolated_list)
```

```python
# Setting the overall aesthetic.
sns.set(style="whitegrid")

ax = sns.boxplot(x='STATE',y='TOTAL_REVENUE',data=CA_TX_isolated,palette='pastel')  

plt.title('Plot 4: Meal Tips by Time and Gender')

sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tips')

plt.show()
```

Pros: Strongly visulizes interquartile ranges, Can view potential outliers
Cons: Lack of context about individual datapoints

```python
sns.set(style="darkgrid")
h = sns.factorplot( x='STATE', y="TOTAL_REVENUE", data=CA_TX_isolated, size=4, kind="bar", palette="pastel", ci=95)
plt.title('Total Revenue of California Schools')
h.set_axis_labels("State", "Total Revenue")
plt.show()
```

Pros: Strongly visulizes means and errors, accurate portrayal of heights of bars
Cons: Lack of context about individual datapoints

```python
sns.set(style="darkgrid")
h = sns.factorplot( x='STATE', y="TOTAL_REVENUE", data=CA_TX_isolated, size=4, kind="point", palette="pastel", ci=95)
plt.title('Total Revenue of California Schools')
h.set_axis_labels("State", "Total Revenue")
plt.show()
```

Pros: Strongly visulizes means and errors
Cons: Lack of context about individual datapoints, potentially misleading in terms of how different the revenues of each state's schools are

```python
h = sns.FacetGrid(CA_TX_isolated, height=5)
sns.lineplot(x="YEAR", y="TOTAL_REVENUE",  data=CA_isolated)
sns.lineplot(x="YEAR", y="TOTAL_REVENUE",  data=TX_isolated)
h.set_axis_labels("Year", "Total Revenue")
plt.title('CA Total Revenue vs. TX Total Revenue')
sns.despine(trim=True)
plt.show()
```

Pros: Shows how individual datapoints relate to eachother over time, easy to read
Cons: Lack of descriptive statistic visualization

```python
h = sns.FacetGrid(CA_TX_isolated, height=4)
sns.distplot(CA_isolated['TOTAL_REVENUE'].dropna())
sns.distplot(TX_isolated['TOTAL_REVENUE'].dropna())
h.set_axis_labels("Total Revenue", "Frequency")
plt.title('CA Total Revenue vs. TX Total Revenue')
sns.despine(trim=True)
plt.show()
```

Pros: Can see where datapoints for each variable cluster 
Cons: Hard to read, Lack of context about individual datapoints

```python
h = sns.FacetGrid(CA_TX_isolated, height=4)
sns.scatterplot(x="YEAR", y="TOTAL_REVENUE",  data=CA_isolated)
sns.scatterplot(x="YEAR", y="TOTAL_REVENUE",  data=TX_isolated)
h.set_axis_labels("Year", "Total Revenue")
plt.title('CA Total Revenue vs. TX Total Revenue')
sns.despine(trim=True)
plt.show()
```

Pros: Shows relative datapoints over time
Cons: Doesn't visually convey the continuous nature of the data, lacks visualization of relationship bewtwwn two dataponints
